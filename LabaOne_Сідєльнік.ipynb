```python
import math

def task_10():
    N = int(input("Введіть число: "))
    
    found = False
    for a in range(1, int(math.sqrt(N)) + 1):
        b_squared = N - a**2
        if b_squared >= 0:
            b = int(math.sqrt(b_squared))
            if b**2 == b_squared:
                print(f'{a}^2 + {b}^2 = {N}')
                found = True
                break
                
    if not found:
        print(f'Число {N} не можна представити квадратами цілих чисел')
```
